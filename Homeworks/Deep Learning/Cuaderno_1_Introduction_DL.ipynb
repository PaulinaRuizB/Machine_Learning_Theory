Librerías

```python
from IPython.display import clear_output
from IPython.display import display
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
import pandas as pd
import numpy as np
import time
```

Exportación de la base de datos

```python
fashion_mnist = keras.datasets.fashion_mnist #cargar fashion mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_valid, X_train = X_train_full[:5000] / 255., X_train_full[5000:] / 255.
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
X_test = X_test / 255.
```

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="softmax")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size = 10,
                    validation_data=(X_valid, y_valid))
```

```python
model.evaluate(X_test, y_test)
```

```python
def build_compile_model():
    model = keras.models.Sequential([
        keras.layers.Flatten(input_shape=[28, 28]),
        keras.layers.Dense(300, activation="relu"),
        keras.layers.Dense(100, activation="relu"),
        keras.layers.Dense(10, activation="softmax")
    ])
    model.compile(loss="sparse_categorical_crossentropy",
                  optimizer="sgd",
                  metrics=["accuracy"])
    return model

print("Function `build_compile_model` defined successfully.")
```

```python
batch_sizes_to_test = [32, 64, 128, 256, 512]
results = []
epochs = 10

for batch_size in batch_sizes_to_test:

    print(f"Training model with batch size: {batch_size}")

    # b. Call the build_compile_model() function to get a new, un-trained model
    model = build_compile_model()

    # c. Record the start time
    start_time = time.time()

    # d. Train the model
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size,
                        validation_data=(X_valid, y_valid), verbose=0)

    # e. Record the end time
    end_time = time.time()

    # f. Calculate total training time
    total_training_time = end_time - start_time

    # g. Calculate average training time per epoch
    avg_time_per_epoch = total_training_time / epochs

    print(f"Finished training for batch size {batch_size}. Evaluating...")

    # h. Evaluate the trained model on the test set
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

    results.append({
        'Batch Size': batch_size,
        'Test Loss': test_loss,
        'Test Accuracy': test_accuracy,
        'Avg Time per Epoch (s)': avg_time_per_epoch
    })

# 4. Clear output after the loop finishes
clear_output(wait=True)
print("Training and evaluation for all batch sizes complete.")
```

```python
df_results = pd.DataFrame(results)
print("Summary Table of Model Performance for different Batch Sizes:")
display(df_results)
```
